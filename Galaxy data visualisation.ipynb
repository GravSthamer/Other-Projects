```python
import matplotlib as mplt
import matplotlib.pyplot as plt
import numpy as np
from numpy import linspace,sin,cos
from numpy import pi
import pandas as pd
from mpl_toolkits.mplot3d import Axes3D
```

# a) Conversion function from spherical to                        cartesian co-ordinates

```python
def cartesian_func(p,theta,phi):
    
    x = p*(sin(phi))*(cos(theta))
    y = p*(sin(phi))*(sin(theta))
    z = p*cos(phi)
   
    return(x,y,z)
```

# b) 

```python
#Provided arrays
phi = np.linspace(-2*pi,2*pi,100)
p = np.linspace(-1.,1.,100)*cos(phi)
theta = np.linspace(-2*pi,2*pi,100)
```

```python
#Calling on conversion function
x,y,z = cartesian_func(p,theta,phi);
```

```python
#Labeling of axis
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.grid()
#Plotting of graph, X,Y.
plt.plot(x,y, c='c')
plt.scatter(x,y,s=30,color='m')
plt.show()
```

# cii) Converting from degrees to radians, and using this new data in my function, then plotting a 3D plot chart of the galaxies.
